```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib.colors import ListedColormap
import xml.etree.ElementTree as ET  
from sklearn.model_selection import KFold
```

```python
# Using XML instead of XLS due to extra 0s in some of the xls files
# Parses XML and returns a dict with {type: total occurrences of that type}
def NPparseLocXML(imageXMLPath: str) -> dict:
    tree = ET.parse(imageXMLPath)
    root = tree.getroot()
    marker_data = root[1][1:]
    
    type_loc = {}
    for marker_type in marker_data:
        typeName = marker_type[0].text
        numTypes = len(marker_type[1:])
        if typeName in type_loc:
            type_loc[typeName] += numTypes
        else:
            type_loc[typeName] = numTypes
    
    return type_loc

# Returns xml file name that can be compared with a jpg or xls file
# Gets rid of suffix .xml, cell counter and counter window if it has those prefixes
# Can remove .xml extension and add wanted extension (etc. ext='xls')
def cleanXMLName(xml, ext=None)->str:
    cc = 'CellCounter_'
    cw = 'Counter Window - '
    xml = xml[:-4]
    
    if (ext is not None):
        xml += f'.{ext}'
        
    if (xml[:len(cc)] == cc):
        xml = xml[len(cc):]
        if (xml[:len(cw)] == cw):
            xml = xml[len(cw):]
    return xml
```

# Cropped and Regular data

```python
testPath = './ExpPatch-Pics/ExpPatchPics-Processed'

# Don't include animation directory 
picDirs = os.listdir(testPath)[1:]

# Gets the counts for each type for each survey
SurveyCountsXLM = {}
# Gets total counts for each patch throughout the surveys
SurveryPatchXLM = {}

# All files in format "Directory (survey) name~Patch Name"
spData = {}

# Count the types
# Open each survey directory
for picDir in picDirs:
    # read all files in directory that are xls to get the count
    surveyXLM = {}
    patchXLM = {}
    
    regPath = os.listdir(f"{testPath}/{picDir}") 
    croppedPath = os.listdir(f"{testPath}/{picDir}/Cropped")
    bothPaths = regPath + croppedPath
    
    regJpgs_xls = [file[:-4] for file in regPath if file.lower().endswith('.jpg') or file.endswith('.xls')]
    croppedJpgs_xls = [file[:-4] for file in croppedPath if file.lower().endswith('.jpg') or file.endswith('.xls')]
    jpg_xls_Names = regJpgs_xls + croppedJpgs_xls
    
    # Parse regular path (not cropped)
    for file in bothPaths:
        xlmName = cleanXMLName(file)
        # only analyze xml if it has corresponding jpg
        if file.endswith('.xml') and xlmName in jpg_xls_Names:
            xlmPath = f'{testPath}/{picDir}/{file}'
            if file in croppedPath:
                xlmPath = f'{testPath}/{picDir}/Cropped/{file}'
                
            xlmDict = NPparseLocXML(xlmPath)
            spData[f'{picDir}~{xlmName}'] = xlmDict
            for specType in xlmDict:
                if specType in surveyXLM:
                    surveyXLM[specType] += xlmDict[specType]
                else:
                    surveyXLM[specType] = xlmDict[specType]
           
            xlmPatchName = xlmName[xlmName.index('YB-')+3:]
            patchXLM[xlmPatchName] = sum(xlmDict.values())
            
    
   
            
                        
    # save types for current survey dir
    SurveyCountsXLM[picDir] = surveyXLM
    SurveryPatchXLM[picDir] = patchXLM
```

# Convert to dataframes

```python
# Counts of each type for each Survery
scx_df = pd.DataFrame(SurveyCountsXLM)

# Get total counts for each type across all survey
psx_counts = pd.DataFrame(scx_df.sum(axis=1), columns=['Counts'])

# Patch total counts for each survey
spx_df = pd.DataFrame(SurveryPatchXLM)
```

# Visualize Types throughout Surveys

```python
# Type counts at each survey
sns.set(font_scale=.6)
ax = sns.heatmap(scx_df, square=True, annot=True)
for t in ax.texts:
    txt = float(t.get_text())
    if txt ==0:
        t.set_text(t.get_text()) 
    elif txt < 50:
        t.set_text("-") 
    elif txt >= 50 and txt < 100:
        t.set_text("*") 
    elif txt >= 100 and txt < 500:
        t.set_text("=") 
    elif txt >= 500 and txt < 1000:
        t.set_text("<") 
    elif txt >= 1000 and txt < 5000:
        t.set_text(">") 
    elif txt >= 5000 and txt < 10000:
        t.set_text("^") 
    elif txt >= 10000 and txt < 20000:
        t.set_text(".") 
    else:
        t.set_text("")
plt.title('Type Counts at each Survey', fontsize=13)
plt.yticks(rotation=0) 
plt.show() 
        
    
# Type counts relative to total survey count
normd_cbs_df = scx_df.div(scx_df.sum(axis=0), axis=1)
sns.set(font_scale=.6)
ax = sns.heatmap(normd_cbs_df, square=True, annot=True)
for t in ax.texts:
    txt = float(t.get_text())
    if txt ==0:
        t.set_text(t.get_text()) 
    else:
        t.set_text("") 
        
plt.title('Type Count Frequency Relative to Surveys', fontsize=13)
plt.yticks(rotation=0) 
plt.show()
```
